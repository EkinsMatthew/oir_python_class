# Part 1: `git`

## What is `git`?

**`git`** is a **version control system** — software that tracks every change
ever made to a set of files.

Think of it like the "make suggestions" mode Google docs, but for entire projects
and much more powerful.

A few key terms:
- **Repository (repo)** — a project folder that `git` is tracking. It contains all the code *and* the full history of every change ever made.
  - This folder that contains the lesson is a repo!
- **Commit** — a saved snapshot of the project at a point in time (like
  hitting "save" on the whole project)
- **Clone** — downloading a complete copy of a repository from a remote
  location (e.g., GitHub) to your own machine

We won't go deep on `git` today — the one command we need is `git clone`.

We already used it, but we will try it out in another context in a moment.

# Part 2: `uv`

## What is `uv`?

**`uv`** is a modern Python package manager — it does roughly the same job
as `conda`, but it is:
- **Much faster** (written in Rust, not Python)
- **Simpler** — one tool handles packages, virtual environments, and even
  Python versions
- **Standard** — built around `pyproject.toml`, the Python community's
  standard format for describing a project

`uv` is a much sleeker and more modern tool.

***

## `uv` vs `conda` — Quick Comparison

| Task                              | `conda`                                 | `uv`                              |
|-----------------------------------|-----------------------------------------|-----------------------------------|
| Install a package                 | `conda install pandas`                  | `uv add pandas`                   |
| Install all project dependencies  | `conda env create -f environment.yml`   | `uv sync`                         |
| Create a new environment          | `conda create -n myenv python=3.11`     | automatic with `uv sync`          |

The biggest difference for today: `uv` stores all installed packages locally, and it maintains a list of instructions that describe exactly how to recreate that environment. Those instructions are stored in the **`pyproject.toml`**.

# Part 3: `pyproject.toml` — The Dependency Manifest

When someone shares a Python project, they need to tell you *what packages
that project needs to run*.

In the `uv` world, this information lives in a file called **`pyproject.toml`**
at the root of the project. A typical one looks like this:

```toml
[project]
name = "my-analysis"
version = "0.1.0"
requires-python = ">=3.11"
dependencies = [
    "pandas>=2.0",
    "matplotlib",
    "openpyxl",
]
```

- **`dependencies`** — the list of packages this project needs
- **`requires-python`** — which Python version to use

When you run `uv sync`, it reads this file and installs everything listed. We don't need to run any manual `pip install` or `conda install` commands.

### Why did that fail?

- Python may not even be installed generically
- The script imports packages that are **not installed** in our current
  Python environment
- Just because the code is on our machine doesn't mean its *dependencies* are
- The `pyproject.toml` tells us exactly what's needed, but we haven't
  installed anything yet

Let's look at what's listed:

View the contents of pyproject.toml. Either use the following command from the terminal
```bash
cat pyproject.toml
```

Or open it with vscode / another text editor

Look at the `dependencies = [...]` section — those are the packages the
script needs.

***

## Exercise 4 — `uv sync`

`uv sync` will:
1. Read `pyproject.toml`
2. Create a **virtual environment** (a self-contained folder called `.venv`)
   inside the project
3. Install all listed dependencies into that environment
4. AND it will do it all blazingly fast

A **virtual environment** is an isolated Python installation that belongs to
this project only; it won't interfere with anything else on your machine.

Run this command:
```bash
uv sync
```

You will see `uv` download and install the packages. This may take a moment
the first time, but `uv` caches packages locally so subsequent installs are
nearly instant.

***

## Exercise 5 — Run the Script with `uv run`

Now that the environment exists, we use **`uv run`** to execute our script
*inside* that environment.

`uv run script.py` is equivalent to activating the environment and
then running Python; it handles that for us automatically.

Run the provided test script with 
```bash
uv run test_script.py
```

It should work now.

This is the core workflow you will encounter when working with shared Python
projects:

```
git clone  ->  uv sync  ->  uv run
```

If you want to get fancy, you can skip the `uv sync` step entirely! 

UV knows that the dependencies are not installed when a new library is loaded, and as a result attempting to run it in UV will install all needed packages as well.

# Summary & Cheatsheet

## The Workflow

| Step | Command                     | What it does                                |
|------|-----------------------------|---------------------------------------------|
| 1    | `git clone <url>`           | Downloads the repo to your machine          |
| 2    | `cd <folder>`               | Navigates into the project directory        |
| 3    | `uv sync`                   | Reads `pyproject.toml` and installs everything |
| 4    | `uv run python script.py`   | Runs the script inside the project environment |

***

## `uv` Reference

| Command                       | What it does                                      |
|-------------------------------|---------------------------------------------------|
| `uv init`                     | Initialize a new project with uv                  |
| `uv sync`                     | Install all dependencies from `pyproject.toml`    |
| `uv add [package_name]`       | Add a new package to the project                  |
| `uv remove [package_name]`    | Remove a package from the project                 |
| `uv run [script].py`          | Run a script in the project environment           |
| `uv pip list`                 | List all installed packages in the environment    |
| `uv --version`                | Check which version of `uv` is installed          |

***

## `git` Reference (the basics)

| Command           | What it does                              |
|-------------------|-------------------------------------------|
| `git clone <url>` | Download a repository from GitHub         |
| `git status`      | See what files have changed               |
| `git log`         | See the history of commits                |
| `git pull`        | Pull down the latest changes from GitHub  |
